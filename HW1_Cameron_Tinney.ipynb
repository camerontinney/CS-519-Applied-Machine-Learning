```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# Importing txt file from desktop into notebook
df=pd.read_csv("C:\\Users\\camer\\OneDrive\\Desktop\\iris.data.txt")
```

```python
# getting number of rows and columns in the data frame iris.data
print("number of rows and columns in df",df.shape)
```

```python
# txt file doesn't have any column headers, we will properly label them here 
df.columns = ['sepal_length', 'sepal_width','petal_length','petal_width','class']
```

```python
# looking at all values in last column "class", print unique values in column. 
print(df["class"].unique())
```

```python
# number of rows that have value "Iris-setosa in column "class" 
print(sum(df['class'] == 'Iris-setosa'))
```

```python
#Average of sepal length by class
df.loc[df["class"] == 'Iris-setosa', 'sepal_length'].mean()
```

```python
#Max of sepal width by Class
df.loc[df["class"] == 'Iris-setosa', 'sepal_width'].max()
```

```python
# Min of petal length by class 
df.loc[df["class"] == 'Iris-setosa', 'petal_length'].min()
```

```python
# Creating a collor and a shape mapping 
colors = {'Iris-setosa':'red', 'Iris-virginica':'blue', 'Iris-versicolor':'green'}
shapes = {'Iris-setosa':'o', 'Iris-virginica':'^', 'Iris-versicolor':'s'}
# Using a for loop to draw the scatter plot
for group, df_group in df.groupby("class"):
    plt.scatter(df_group['sepal_length'], df_group['sepal_width'],
color=colors[group], marker=shapes[group], label = group)

# adding axis labels and legend
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.legend()

# show plot
plt.show()
```
